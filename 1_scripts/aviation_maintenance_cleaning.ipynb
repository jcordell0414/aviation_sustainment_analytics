ADDRESS THE NULL VALUES IN THE DATASET

```python
#Display the unique values of the supplier delay days column
supplier_delay_values = df['supplier_delay_days'].unique()
print(supplier_delay_values)
```

```python
#Convert valid numeric values to numbers and nan to NaN to be handled
df['supplier_delay_days'] =pd.to_numeric(df['supplier_delay_days'], errors = 'coerce')
#print(df['supplier_delay_days'].median())
```

```python
#Replace NULL values with the median of the supplier delay days column
df['supplier_delay_days'] = df['supplier_delay_days'].fillna(df['supplier_delay_days'].median())
```

```python
#Check to see if NULL values have been filled
supplier_delay_values = df['supplier_delay_days'].unique()
print(supplier_delay_values)
```

```python
#Repeat the process for the downtime hours column
downtime_hours_values = df['downtime_hours'].unique()
print(downtime_hours_values)
```

```python
df['downtime_hours'] = pd.to_numeric(df['downtime_hours'],errors = 'coerce')
df['downtime_hours'] = df['downtime_hours'].fillna(df['downtime_hours'].median())
```

```python
downtime_hours_values = df['downtime_hours'].unique()
print(downtime_hours_values)
```

```python
#Run a NULL check again
print('\n=== NULL COUNTS ===')
print(df.isnull().sum())
```

NORMALIZE ALL THE DATES INTO THE SAME FORMAT (MM-DD-YYYY)

```python
#df['event_date'] = pd.to_datetime(df['event_date'], errors = 'coerce')
```

```python
df = df.copy()
```

```python
#Convert all values in event date column to strings for parsing and normalization
df['event_date'] = df['event_date'].astype(str)
```

```python
#Find all the YYYY-MM-DD dates in event date column
mask = df['event_date'].str.match(r'^\d{4}-\d{2}-\d{2}$')

#Convert these dates to MM-DD-YYYY and datetimes
df.loc[mask, 'event_date'] = pd.to_datetime(df.loc[mask, 'event_date'], format = '%Y-%m-%d').dt.strftime('%m-%d-%Y')
```

```python
df.head()
```

```python
#Replace all / with - in event date column
df['event_date'] = df['event_date'].str.replace('/','-')
df.head()
```

```python
#Finally convert full event date column to datetimes
df['event_date'] = pd.to_datetime(df['event_date'], format = '%m-%d-%Y', errors='coerce')
#Verify the ype conversion
df.info()
```

```python
#Verify no nulls
df.isnull().sum()
```

DATA HAS BEEN CLEANED
NOW CREATE NEW COLUMNS FOR TABLEAU VISUALIZATION USE

```python
#Create a month and year column
df['month'] = df['event_date'].dt.month
df['year'] = df['event_date'].dt.year
```

```python
#Create a total delay column
df['total_delay'] = df['downtime_hours']+df['supplier_delay_days']
```

```python
#Convert readiness score to numeric then Create a risk category column
df['readiness_score'] = pd.to_numeric(df['readiness_score'])
df['risk_level'] = np.where(df['readiness_score']<70,'High Risk', 'Normal')
```

INITIAL ANALYSIS

```python
#Average down time by aircraft
df.groupby('aircraft_model')['downtime_hours'].mean()
```

```python
#Highest Supplier Delays
df.groupby('supplier_name')['supplier_delay_days'].mean().sort_values(ascending=False)
```

```python
#Maintenance Type with Highest down times
df.groupby('maintenance_type')['downtime_hours'].mean().sort_values(ascending=False)
```

```python
df.info()
```

```python
#Convert maintenance cost column from string to numeric
df['maintenance_cost'] = pd.to_numeric(df['maintenance_cost'])
```

KPI ANALYSIS

```python
#Get the mean of the down time hours and maintenance cost columns as well as count of the maintenance ids
#Grouped by maintenance type
#Rounded to 2 decimal places
df.groupby('maintenance_type').agg({
    'downtime_hours':'mean', 
    'maintenance_cost':'mean', 
    'maintenance_id':'count'
}).round(2).sort_values(by='downtime_hours', ascending = False)
```
